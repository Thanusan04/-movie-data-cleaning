```python
import pandas as pd
```

```python
movie=pd.read_csv("movies.csv")
movie
```

```python
movie.isnull()
```

```python
movie.isnull().sum()
```

```python
movie.info()
```

```python
text_columns = ['MOVIES', 'GENRE', 'ONE-LINE', 'STARS']

for col in text_columns:
    movie[col] = movie[col].astype(str).str.replace('\n', '', regex=True).str.strip()
```

```python
movie
```

```python
movie['YEAR'] = movie['YEAR'].astype(str).str.extract(r'(\d{4})')
```

```python
movie
```

```python
movie['VOTES'] = movie['VOTES'].str.replace(',', '', regex=False)
movie['VOTES'] = pd.to_numeric(movie['VOTES'], errors='coerce')
```

```python
movie
```

```python
movie.drop(columns=['Gross'], inplace=True)
```

```python
movie
```

```python
movie.dropna(subset=['RATING', 'RunTime'], inplace=True)
```
